```python
df = pd.read_csv('mtr_meter.csv', sep=';')
df['start_time'] = pd.to_datetime(df.start_time)
df_f = df[df.hop != 2]
```

```python
d_bkp = ColumnDataSource(df_f)
```

```python
p = bkp.figure(x_axis_type='datetime', plot_width=1200)
g = p.line('start_time', 'avg', source=d_bkp, line_width=3)
s = p.asterisk('start_time', 'avg', source=d_bkp, size=6)
bkp.show(p)
```
